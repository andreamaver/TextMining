```python
import json
import csv
import pandas as pd
import gzip
from google.colab import drive
from shutil import copyfile
import tarfile

import random 
random.seed(10)
```

```python
# Link Google Drive account for dataset
drive.mount('/content/gdrive')

# Copy file in Google Colab Environment
copyfile('gdrive/My Drive/Progetto_TMeS/bigPatentData/train.tar.gz', 'train.tar.gz')

# Open file
file = tarfile.open('train.tar.gz')

# Extracting file
file.extractall('./train_patent')
file.close()
```

```python
#Select Topic and the number of document to unzip
topic=['a', 'b', 'c', 'd', 'e','f', 'g', 'h']
range_num=['0','1','2','3']
```

```python
for r in range_num:
  for t in topic:
    # Generic path for file in train_patent
    path='train_patent/train/'+t+'/data00000000000'+r+'.gz'

    # Unzip file
    t_file=gzip.open(path, "rb")
    orig=t_file.read()
    data = orig.decode('utf-8').splitlines()

    # Create empty dictionary
    l=len(data)
    dic=['']*l

    # Value of dictionary are data
    for i in range(0,l):
      dic[i]=json.loads(data[i])

    # Create csv for each unzip file
    name='train_'+t+'_'+r+'.csv'
  
    with open(name, 'w') as f:
      for i in range(0,len(dic)):
        w = csv.DictWriter(f, dic[i].keys())
        if i==0:
          w.writeheader()
        w.writerow(dic[i])
```

```python
# Read first file
data=pd.read_csv('train_a_0.csv')
# Assign topic 
data['topic']='a'
for r in range_num:
  for t in topic:
    file_tmp='train_'+t+'_'+r+'.csv'
    if (file_tmp!='train_a_0.csv'):
      # Read each csv file
      tmp=pd.read_csv(file_tmp)
      tmp['topic']=t
      frames=[data, tmp]
      # Concat them
      data=pd.concat(frames)
```

```python
data=data[['abstract','description','topic']]
```

```python
# Shuffle data
data=data.sample(frac=1).reset_index(drop=True)
```

```python
# Save csv
data.to_csv('/content/gdrive/MyDrive/Progetto_TMeS/brevetti_class.csv')
```
